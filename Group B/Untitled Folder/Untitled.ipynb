### Create a review scrapper for any ecommerce website to fetch real time comments, reviews, ratings, comment tags, customer name using Python.



```python
import requests
import bs4
```

```python
request1 = requests.get("https://www.flipkart.com/apple-iphone-14-plus-purple-128-gb/p/itm7a759a472d863?pid=MOBGHWFHYFJWRHAG&lid=LSTMOBGHWFHYFJWRHAGX1JLRZ&marketplace=FLIPKART&q=iphone+14+plus&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&fm=personalisedRecommendation%2FC5&iid=54c6fb73-dff3-41f0-9155-6b2b1af1d875.MOBGHWFHYFJWRHAG.SEARCH&ppt=pp&ppn=pp&ssid=gqbscw22wu0kderk1684987839121&qH=70897fd4f67467a5")
```

```python
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
reviews = soup.find_all('div',{'class':'t-ZTKy'});

for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
ratings =soup.find('div',{'class': "_3LWZlK _1BLPMq"}).get_text();
ratings
```

```python
ratings = soup.find_all('div',{'class':"_3LWZlK"});
for ind_rating in ratings:
    print(ind_rating.get_text() + "\n")
```

```python
customer_name = soup.find_all("p", {"class":"_2sc7ZR _2V5EHH"});

for cust_name in customer_name:
    print(cust_name.get_text() + "\n")
```
