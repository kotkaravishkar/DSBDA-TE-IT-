### Facebook Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
%matplotlib inline
```

```python
df=pd.read_csv("facebook.csv", sep=";")
```

```python
df
```

```python
df.head()
```

```python
post_types=df.Type.unique()
post_types
```

```python
# Likes cound in Dataset

likes = {}

for post in post_types:
    subset = df[df.Type == post]
    likes[post] = subset.like.sum()
    
likes
```

```python
fig = plt.figure(figsize=(8, 8))

# Adds subplot on position 1
ax = fig.add_subplot(111)

# Generating and displaying bar chart
plt.bar(
    x=likes.keys(),
    height=likes.values()
)
plt.xlabel("Type of Post", fontsize=20)
plt.ylabel("Number of Likes", fontsize=20)
plt.title("Number of likes per type (Bar Plot)", fontsize=20)
plt.show()
```

```python
#frequency data for each type

frequency= {}
for post in post_types:
    subset = df[df.Type == post]
    frequency[post] = subset.shape[0]

frequency
```

```python
frequency
```

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111)


legend = [
    "Photo",
    "Status",
    "link",
    "Video"
]

explode = [0.1,0.1,0.1,0.1]

plt.pie(
    x=frequency.values(),
    labels=legend,
    explode=explode,
)
```

#### Paid and Unpaid Posts

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111)

sns.countplot(x=df.Paid)

plt.xlabel("Paid Posts")
plt.ylabel("Count")
plt.title("Paid vs UnPaid")
```

### 2. Heart Disease dataset

```python
heart_data=pd.read_csv("heart.csv")
```

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111)

sns.boxplot(x=heart_data.thalach)
plt.title("Quartile spread of thalach feature (Box plot)", fontsize=20)

plt.show()
```
